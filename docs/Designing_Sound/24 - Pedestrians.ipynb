# Pedestrians

```python
from synthingie import Audio, Module
SAMPLERATE = 48000
FRAMESIZE = 1024
```

## This is the sound to emulate

```python
Audio.load("pedestrian-beeps.wav").display()
```

## This is the oscillator to gate

```python
mod = Module(SAMPLERATE, FRAMESIZE)
osc = mod.sin(2500, 0.2)

mod.render(osc, 10).display()
```

## This is the gate

```python
gate = mod.naive_square(5, amplitude=0.5) + 0.5

mod.render(gate, 10).display(limit_samples=48000)
```

## This is everything together

```python
pedestrian = osc * gate

mod.render(pedestrian, 5).display()
```
