```python
from synthingie import Module
```

# Line: Generate ramps

```python
mod = Module(48000, 1024)
line = mod.line()
line.set(
    (1, 1),
    (0, 2),
    (0.5, 3),
    (0, 0.5),
)
mod.render(line, 10).display_signal()
```

# ADSR Envelopes

```python
mod = Module(48000, 1024)
env = mod.adsr(1, 1, 0.5, 0.5, 2)
env.on()
mod.render(env, 5).display_signal()
```

```python
env.off()
mod.render(env, 5).display_signal()
```

# With scores

```python
from synthingie import score

mod = Module(48000, 1024)
env = mod.adsr(1, 1, 0.5, 0.5, 2)

sc = score.Call(env.on) + score.Wait(3) + score.Call(env.off)

mod.render(env, 5, score=sc).display_signal()
```
