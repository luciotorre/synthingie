```python
import synthingie as st
```

```python
wave = (
    st.Sin(
        st.Parameter(400, name="freq", min=2, max=1200),
        amplitude = 0.5
    )
).lop(
    st.Parameter(1000, name="low", min=2, max=1200), 
    st.Parameter(1, name="low-q", min=2, max=50), 
).hip(
    st.Parameter(1, name="hi", min=2, max=1200),
    st.Parameter(1, name="hi-q", min=2, max=50), 
)

wave.live(controls=[
    st.live.Analyzer(wave),
    st.live.Scope(wave)
])
```
