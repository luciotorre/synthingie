```python
from synthingie import Module
from contextlib import contextmanager
```

```python
SAMPLERATE = 44100
FRAMESIZE = 441


@contextmanager
def viewer():
    mod = Module(SAMPLERATE, FRAMESIZE)    
    yield mod
    target = mod._steps[-1]
    audio = mod.render(target, duration_s=10, profile=True)
    audio.display(limit_samples=1000)
```

```python
with viewer() as mod:
    wave = mod.sin(440)
```

```python
with viewer() as mod:
    wave = mod.analog_sin(440)
```

```python
with viewer() as mod:
    wave = mod.naive_saw(440)
```

```python
with viewer() as mod:
    wave = mod.saw(440)
```

```python
with viewer() as mod:
    wave = mod.analog_saw(440)
```

```python
with viewer() as mod:
    wave = mod.naive_triangle(440)
```

```python
with viewer() as mod:
    wave = mod.triangle(440)
```

```python
with viewer() as mod:
    wave = mod.naive_square(440)
```

```python
with viewer() as mod:
    wave = mod.square(440)
```
