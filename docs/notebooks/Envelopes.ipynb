```python
import synthingie as st
```

# Line: Generate ramps

```python
line = st.Line()
line.set(
    (1, 1),
    (0, 2),
    (0.5, 3),
    (0, 0.5),
)
line.render(10).display_signal()
```

# ADSR Envelopes

```python
env = st.ADSR(1, 1, 0.5, 0.5, 2)
env.on()
env.render(5).display_signal()
```

```python
env.off()
env.render(5).display_signal()
```

# With scores

```python
env = st.ADSR(1, 1, 0.5, 0.5, 2)

sc = st.score.Call(env.on) + st.score.Wait(3) + st.score.Call(env.off)

env.render(5, score=sc).display_signal()
```
